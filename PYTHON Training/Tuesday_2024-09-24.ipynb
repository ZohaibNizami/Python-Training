```python
import pandas as pd
import json
# Read the CSV file into a pandas DataFrame
file_path = 'Task_1.csv'
df = pd.read_csv(file_path)
# Function to split the full name into first, middle, and last names (simplified)
   #split a string into a list where each word is a list item:
def split_name(full_name):
    name_parts = full_name.split()
    if len(name_parts) == 2:
        return name_parts[0], "", name_parts[1]
    elif len(name_parts) == 3:
        return name_parts[0], name_parts[1], name_parts[2]
    return full_name, "", ""
# Transform the DataFrame into the desired format
def transform_row(row):
    first_name, middle_name, last_name = split_name(row['Patient Name'])
    transformed_data = {
        "account_number": row['Patient ID'],
        "service_date": row['Service Date'],
        "debtor": {
            "first_name": first_name,
            "last_name": last_name,
            "middle_name": middle_name,
            "date_of_birth": row['Date of Birth'],
        },
        "responsible_person": {},
        "procedure_code": row['Procedure Code'],
        "procedure_description": row['Procedure Description'],
        "initial_account_balance": row['Billing Amount'],
        "current_account_balance": row['Billing Amount'],
        "amount_paid_to_date": 0,  # Assuming no payments made yet
        "insurance": {
            "insurance_provider": row['Insurance Provider'],
            "policy_number": row['Policy Number']
        }
    }
    return transformed_data
# Apply the transformation to all rows in the DataFrame
transformed_data = df.apply(transform_row, axis=1)
# Convert the transformed data into a list of dictionaries
transformed_data_list = transformed_data.tolist()
# Specify the JSON output file path
json_output_file = 'transformed_data.json'
# Save the data to a JSON file
with open(json_output_file, 'w') as json_file:
    json.dump(transformed_data_list, json_file, indent=4)
# Confirm that the file has been saved
print(f'Data has been successfully saved to {json_output_file}')
df
```
